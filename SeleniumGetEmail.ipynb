```python
# get emails 

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import csv
import pandas as pd
import time

driver = webdriver.Chrome(ChromeDriverManager().install())

driver.get('https://thedatapack.com/tools/find-github-user-email/')


inputElement = driver.find_element_by_id('username')

df = pd.DataFrame()

f = open("/media/johnatan/SAMSUNG/toSend/filens.txt", "r")

for nameDev in f:
    
    try:
        
        inputElement.send_keys(nameDev)
        inputElement.send_keys(u'\ue007')

        time.sleep(15)

        email = driver.find_element_by_xpath('/html/body/div[2]/div/div/div[2]/div[2]/div[2]/div[2]').text

        time.sleep(15)

        name = driver.find_element_by_xpath('/html/body/div[2]/div/div/div[2]/div[1]/div[2]').text
        gituser=driver.find_element_by_xpath('/html/body/div[2]/div/div/div[2]/div[1]/div[3]/span[2]/a').text
        print(name,';',email,';',gituser)
        df = df.append({'name':name,'e-mail':email}, ignore_index=True)
        driver.find_element_by_id('username').clear()
        
    except:
        driver.find_element_by_id('username').clear()
        
        print("An exception occurred; exception; occurred")

df.to_csv("/media/johnatan/SAMSUNG/emailDev.csv")
```

```python
#get name of developer to clone 

import pandas as pd
import time 

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import csv


driver = webdriver.Chrome(ChromeDriverManager().install())

df = pd.DataFrame()

f = open("/media/johnatan/SAMSUNG/SANER2021/linksGit.txt", "r")

dflinks = pd.DataFrame()


for contributors in f:
    driver.get(contributors)
    time.sleep(5)
    
    content_blocks = driver.find_elements_by_id("contributors")
    try:
        for block in content_blocks:
            elements = block.find_elements_by_tag_name("a")
            for el in elements:
                dflinks = dflinks.append({'git':el.get_attribute("href")}, ignore_index=True)
                #print(el.get_attribute("href"))
    except:
        print("An exception occurred "+contributors)
        
dflinks
```

```python
dflinks.drop_duplicates(subset ="git",keep = False, inplace = True)

dflinks.to_csv("/media/johnatan/SAMSUNG/SANER2021/emailDevXXXX.csv",index=False)
```
